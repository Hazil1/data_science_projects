### Разделение данных 

```python
# Разделим данные на признаки и целевой признак
features = data['text']
target = data['toxic']

features.head(), target.head()
```

```python
# Разделим датасет на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(features, 
                                                    target,
                                                    test_size=0.2,
                                                    random_state=RANDOM_STATE)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

Создадим функцию для перевода текстов в векторное представление. Будем использовать предобученную модель `Kev07/Toxic-Tweets`

```python
def make_features(data):
    model_path = 'Kev07/Toxic-Tweets'
    tokenizer = transformers.AutoTokenizer.from_pretrained(model_path)
    
    max_len = 512
    tokenized = data.apply(
        lambda x: tokenizer.encode(x, add_special_tokens=True, 
                                   max_length=max_len, 
                                   truncation=True,))

    for i in tokenized.values:
        if len(i) > max_len:
            max_len = len(i)

    padded = np.array([i + [0]*(max_len - len(i)) for i in tokenized.values])

    attention_mask = np.where(padded != 0, 1, 0)

    config = transformers.AutoConfig.from_pretrained(model_path)
    model = transformers.AutoModel.from_pretrained(model_path)

    batch_size = 100
    device = torch.device('mps')
    model.to(device)
    embeddings = []
    
    for i in notebook.tqdm(range(padded.shape[0] // batch_size)):
        batch = torch.LongTensor(padded[batch_size*i:batch_size*(i+1)]).to(device) 
        attention_mask_batch = torch.LongTensor(attention_mask[batch_size*i:batch_size*(i+1)]).to(device)
        
        with torch.no_grad():
            batch_embeddings = model(batch, attention_mask=attention_mask_batch)

        embeddings.append(batch_embeddings[0][:,0,:].cpu().numpy())
        
    return np.concatenate(embeddings)
```

```python
X_train = make_features(X_train)
```

```python
X_test = make_features(X_test)
```

## Обучение

Исходя из задачи нам необходимо решить задачу классификации.

Для выбора наилучшей модели обучим 3 модели:
1. LogisticRegression
2. RandomForestClassifier
3. CatBoostClassifier

### Обучим модель LogisticRegression

```python
lrm = LogisticRegression(random_state=RANDOM_STATE, max_iter=2000)
param_grid_lrm = {
    'class_weight': (None, 'balanced'),
    'solver': ('liblinear', 'lbfgs'),
}

grid_search_lrm = RandomizedSearchCV(estimator=lrm,
                                     param_distributions=param_grid_lrm,
                                     n_jobs=-1,
                                     random_state=RANDOM_STATE,
                                     scoring='f1')

pred_lrm = grid_search_lrm.fit(X_train, y_train)
```

```python
print(f'''
Лучшая оценка F1: {pred_lrm.best_score_}
Лучшие параметры модели: {pred_lrm.best_params_}
''')
```

### Обучим модель RandomForestClassifier

```python
rfm = RandomForestClassifier(random_state=RANDOM_STATE)
param_grid_rfm = {
    'n_estimators': range(10, 101, 10),
    'max_depth': range(1, 11, 1),
    'min_samples_split': range(2, 11, 1)
}
grid_search_rfm = RandomizedSearchCV(estimator=rfm,
                                     param_distributions=param_grid_rfm,
                                     n_jobs=-1,
                                     random_state=RANDOM_STATE,
                                     scoring='f1')

pred_rfm = grid_search_rfm.fit(X_train, y_train)
```

```python
print(f'''
Лучшая оценка F1: {pred_rfm.best_score_}
Лучшие параметры модели: {pred_rfm.best_params_}
''')
```

### Обучим модель CatBoostClassifier

```python
cbr = CatBoostClassifier(random_state=RANDOM_STATE, eval_metric='TotalF1')

param_grid_cbr = {
    'max_depth': range(1, 7, 1),
    'n_estimators': range(10, 110, 10),
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.5, 1, 2, 5],
}

grid_search_cbr = cbr.randomized_search(param_grid_cbr,
                                        X=X_train,
                                        y=y_train,
                                        refit=True,
                                        cv=5)
```

```python
print(f'Лучшая оценка: {cbr.best_score_}')
```

```python
results = {'Model': ['LogisticRegression', 'RandomForestClassifier', 'CatBoostClassifier'],
           'F1': [pred_lrm.best_score_, pred_rfm.best_score_, cbr.best_score_['learn']['TotalF1']]}
results = pd.DataFrame(data=results)

results
```

В результате обучения моделей наилучший результат показала модель `CatBoostClassifier`.

Протестируем модель на тестовой выборке.

```python
predictions_test = cbr.predict(X_test)
score = f1_score(y_test, predictions_test)
print(f'F1 для модели CatBoostClassifier: {score}')
```
